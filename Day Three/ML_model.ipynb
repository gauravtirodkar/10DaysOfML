```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('http://iali.in/datasets/loan_status_train.csv')
```

```python
df.head(20)
```

```python
df.shape
```

```python
#Cleaning Data
df.isnull().sum()
```

```python
print(df['Gender'].value_counts())
print(df['Married'].value_counts())
print(df['Dependents'].value_counts())
print(df['Self_Employed'].value_counts())
print(df['LoanAmount'].value_counts())
print(df['Loan_Amount_Term'].value_counts())
print(df['Credit_History'].value_counts())
```

```python
df.drop(columns=['Loan_ID'],inplace=True)
```

```python
df['Gender'].fillna('Male',inplace=True)
df['Married'].fillna('Yes',inplace=True)
df['Self_Employed'].fillna('No',inplace=True)
df['Credit_History'].fillna('1',inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Gender'] = df['Gender'].map({'Male':1,'Female':0})
df['Married'] = df['Married'].map({'Yes':1,'No':0})
df['Education'] = df['Education'].map({'Graduate':1,'Not Graduate':0})
df['Self_Employed'] = df['Self_Employed'].map({'Yes':1,'No':0})
```

```python
df['Income'] = df['ApplicantIncome'] + df['CoapplicantIncome']
df.drop(['ApplicantIncome','CoapplicantIncome'], axis=1, inplace=True)
```

```python
df['Singleton'] = df['Dependents'].map(lambda d : 1 if d=='1' else 0) 
df['Small_Family'] = df['Dependents'].map(lambda d : 1 if d=='2' else 0) 
df['Large_Family'] = df['Dependents'].map(lambda d : 1 if d=='3+' else 0) 
df.drop(columns=['Dependents'],inplace=True)
```

```python
def feature_scaling(dataframe):
    dataframe -= dataframe.min()
    dataframe /= dataframe.max()
    return dataframe
```

```python
df['LoanAmount'] = feature_scaling(df['LoanAmount'])
df['Income'] = feature_scaling(df['Income'])
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
df.drop(columns=['Property_Area','Loan_Amount_Term'],inplace=True)
df.head()
```

```python
target = df['Loan_Status'][:490].map({'Y':1,'N':0})
train = df.iloc[:490].drop(columns=['Loan_Status'])
test = df.iloc[491:]
train
```

```python
clf = RandomForestClassifier(n_estimators=50,max_features='sqrt')
clf = clf.fit(train,target)
```

```python
np.mean(cross_val_score(clf, train, target, cv = 5, scoring='accuracy'))
```

```python
from sklearn.model_selection import train_test_split
Y = df['Loan_Status'].map({'Y':1,'N':0})
df.drop(columns=['Loan_Status'],inplace=True)
```

```python
train_X,test_X,train_Y,test_Y = train_test_split(df,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(train_X,train_Y)
```

```python
print("Training Score : ",logreg.score(train_X,train_Y))
```

```python
print("Testing Score : ",logreg.score(test_X,test_Y))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(train_X,train_Y)
```

```python
print("Testing Score : ",rfc.score(test_X,test_Y))
```
